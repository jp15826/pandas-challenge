# PyCitySchools Analysis:

       Overall, charter schools with a lower budget per student (~$585) performed higher than district schools with a higher budget per student (~$630).
    
       Overall, small and medium sized (<2000 students) schools performed higher than large sized schools (2000+ students).
    
       Overall, charter schools exceeded district school students' average math and reading scores by approximately 7% and 3% respectively.
    
   - Charter schools out performed district schools with higher average math scores, average reading scores, percentage of students passing math, percentage of students passing reading, and overall passing rate. It is difficult to conclude if these results are due to the fact that charter schools have less students than district schools or if there is another variable affecting this analysis.
    

```python
import pandas as pd

schoolsFile="Resources/schools_complete.csv"
studentFile="Resources/students_complete.csv"

schoolsDF=pd.read_csv(schoolsFile)
studentDF=pd.read_csv(studentFile)


schoolCompleteDF=pd.merge(studentDF,schoolsDF,how="left", on=["school_name", "school_name"])
schoolCompleteDF.head()
```

District Summary

```python
schoolCount=len(schoolCompleteDF["school_name"].unique())
schoolCount
```

```python
studentCount=len(schoolCompleteDF ["Student ID"].unique ())
studentCount
```

```python
totalBudget=schoolsDF["budget"].sum()
totalBudget
```

```python
averageMathScore= schoolCompleteDF["math_score"].mean ()
averageMathScore
```

```python
averageReadingScore= schoolCompleteDF ["reading_score"].mean()
averageReadingScore
```

```python
passingMathCount =  schoolCompleteDF[(schoolCompleteDF["math_score"] >= 70)].count()["student_name"]
passingMathPercentage = passingMathCount / float(studentCount) * 100
passingMathPercentage
```

```python
passingReadingCount = schoolCompleteDF[(schoolCompleteDF["reading_score"] >= 70)].count()["student_name"]
passingReadingPercentage = passingReadingCount / float(studentCount) * 100
passingReadingPercentage
```

```python
passingMathReadingCount = schoolCompleteDF[
    (schoolCompleteDF["math_score"] >= 70) & (schoolCompleteDF["reading_score"] >= 70)
].count()["student_name"]
overallPassingRate = passingMathReadingCount /  float(studentCount) * 100
overallPassingRate
```

```python
districtSummary={
     "Total Schools": [schoolCount],
    "Total Students": [studentCount],
    "Total Budget": [totalBudget],
    "Average Math Score": [averageMathScore],
    "Average Reading Score": [averageReadingScore],
    "% Passing Math": [passingMathCount],
    "% Passing Reading": [passingReadingCount],
    "% Overall Passing": [passingMathReadingCount]
}
districtSummaryDF=pd.DataFrame(districtSummary)
districtSummaryDF["Total Students"]=districtSummaryDF["Total Students"].map("{:,}".format)
districtSummaryDF["Total Budget"]=districtSummaryDF["Total Budget"].map("{:,.2f}".format)
districtSummaryDF
```

School Summary

```python
schoolTypes=schoolsDF.set_index(["school_name"])["type"]
```

```python
perSchoolCounts = schoolCompleteDF.groupby(["school_name"]).count()["size"]
```

```python
perSchoolBudget = schoolCompleteDF.groupby(["school_name"]).mean()["budget"]
perSchoolCapita = perSchoolBudget / perSchoolCounts
```

```python
perSchoolMath = schoolCompleteDF.groupby(["school_name"]).mean()["math_score"]
perSchoolReading = schoolCompleteDF.groupby(["school_name"]).mean()["reading_score"]
```

```python
schoolPassingMath= schoolCompleteDF[(schoolCompleteDF["math_score"] >= 70)]
```

```python
schoolPassingReading =schoolCompleteDF[(schoolCompleteDF["reading_score"] >= 70)]
```

```python
passingMathAndReading = schoolCompleteDF[
    (schoolCompleteDF["reading_score"] >= 70) & (schoolCompleteDF["math_score"] >= 70)
]
```

```python
perSchoolPassingMath = schoolPassingMath.groupby(["school_name"]).count()["student_name"] / perSchoolCounts * 100
perSchoolPassingReading = schoolPassingReading.groupby(["school_name"]).count()["student_name"] / perSchoolCounts * 100
overallPassingRate=passingMathAndReading.groupby(["school_name"]).count()["student_name"] / perSchoolCounts * 100
```

```python
perSchoolSummary={
    "School Type":schoolTypes,
    "Total Students": perSchoolCounts,
    "Total School Budget": perSchoolBudget,
    "Per Student Budget":perSchoolCapita,
    "Average Math Score": perSchoolMath,
    "Average Reading Score": perSchoolReading,
    "% Passing Math": perSchoolPassingMath,
    "% Passing Reading": perSchoolPassingReading,
    "% Overall Passing": overallPassingRate
}
perSchoolSummaryDF=pd.DataFrame(perSchoolSummary)
perSchoolSummaryDF
```

Highest-Performing Schools

```python
highestPerforming=perSchoolSummaryDF.sort_values("% Overall Passing",ascending=False)
highestPerforming.head(5)
```

Bottom Performing Schools

```python
lowestPerforming=perSchoolSummaryDF.sort_values("% Overall Passing",ascending=True)
lowestPerforming.head(5)
```

Math Scores By Grade

```python
ninthGraders = schoolCompleteDF[(schoolCompleteDF["grade"] == "9th")]
tenthGraders = schoolCompleteDF[(schoolCompleteDF["grade"] == "10th")]
eleventhGraders = schoolCompleteDF[(schoolCompleteDF["grade"] == "11th")]
twelfthGraders = schoolCompleteDF[(schoolCompleteDF["grade"] == "12th")]

ninthGradersScores = ninthGraders.groupby(["school_name"]).mean()
tenthGradersScores = tenthGraders.groupby(["school_name"]).mean()
eleventhGradersScores = eleventhGraders.groupby(["school_name"]).mean()
twelfthGradersScores = twelfthGraders.groupby(["school_name"]).mean()

ninthGradeMathScores = ninthGradersScores["math_score"]
tenthGradeMathScores = tenthGradersScores["math_score"]
eleventhGradeMathScores = eleventhGradersScores["math_score"]
twelfthGradeMathScores = twelfthGradersScores["math_score"]


mathScoresByGrade={
    "9th":ninthGradeMathScores,
    "10th":tenthGradeMathScores,
    "11th":eleventhGradeMathScores,
    "12th":twelfthGradeMathScores
}
mathScoresByGradeDF=pd.DataFrame(mathScoresByGrade)
mathScoresByGradeDF.index.name=None
mathScoresByGradeDF
```

Reading Scores By Grade

```python
ninthGraders = schoolCompleteDF[(schoolCompleteDF["grade"] == "9th")]
tenthGraders = schoolCompleteDF[(schoolCompleteDF["grade"] == "10th")]
eleventhGraders = schoolCompleteDF[(schoolCompleteDF["grade"] == "11th")]
twelfthGraders = schoolCompleteDF[(schoolCompleteDF["grade"] == "12th")]

ninthGradersScores = ninthGraders.groupby(["school_name"]).mean()
tenthGradersScores = tenthGraders.groupby(["school_name"]).mean()
eleventhGradersScores = eleventhGraders.groupby(["school_name"]).mean()
twelfthGradersScores = twelfthGraders.groupby(["school_name"]).mean()

ninthGradeReadingScores = ninthGradersScores["reading_score"]
tenthGradeReadingScores = tenthGradersScores["reading_score"]
eleventhGradeReadingScores = eleventhGradersScores["reading_score"]
twelfthGradeReadingScores = twelfthGradersScores["reading_score"]


ReadingScoresByGrade={
    "9th":ninthGradeReadingScores,
    "10th":tenthGradeReadingScores,
    "11th":eleventhGradeReadingScores,
    "12th":twelfthGradeReadingScores
}
ReadingScoresByGradeDF=pd.DataFrame(ReadingScoresByGrade)
ReadingScoresByGradeDF.index.name=None
ReadingScoresByGradeDF
```

Scores by School Spending

```python
spendingBins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
schoolSpendingDF = perSchoolSummaryDF.copy()
```

```python
schoolSpendingDF["Spending Ranges (Per Student)"]=pd.cut(schoolSpendingDF['Per Student Budget'],spendingBins,labels=labels,include_lowest=True)
schoolSpendingDF
```

```python
spendingMathScores = schoolSpendingDF.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spendingReadingScores = schoolSpendingDF.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spendingPassingMath = schoolSpendingDF.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spendingPassingReading = schoolSpendingDF.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overallPassingSpending = schoolSpendingDF.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]
```

```python
spendingSummary={
    "Average Math Score":spendingMathScores,
    "Average Reading Score":spendingReadingScores,
    "% PassingMath":spendingPassingMath,
    "% Passsing Reading":spendingPassingReading,
    "% Overall Passing":overallPassingSpending
}

spendingSummaryDF=pd.DataFrame(spendingSummary)
spendingSummaryDF
```

Scores By School Size

```python
sizeBins=[0,1000,2000,5000]
labels=["Small (<1000)","Medium (1000-2000)","Large (2000-5000)"]
```

```python
schoolSizeDF=perSchoolSummaryDF.copy()
```

```python
schoolSizeDF['School Size']=pd.cut(schoolSizeDF['Total Students'],sizeBins,labels=labels,include_lowest=True)
schoolSizeDF
```

```python
sizeMathScores = schoolSizeDF.groupby(["School Size"]).mean()["Average Math Score"]
sizeReadingScores = schoolSizeDF.groupby(["School Size"]).mean()["Average Reading Score"]
sizePassingMath = schoolSizeDF.groupby(["School Size"]).mean()["% Passing Math"]
sizePassingReading = schoolSizeDF.groupby(["School Size"]).mean()["% Passing Reading"]
sizeOverallPassing = schoolSizeDF.groupby(["School Size"]).mean()["% Overall Passing"]
```

```python
sizeSummary={
    "Average Math Score": sizeMathScores,
    "Average Reading Score": sizeReadingScores,
    "% Passing Math": sizePassingMath,
    "% Passing Reading": sizePassingReading,
    "% Overall Passing": sizeOverallPassing
}

sizeSummaryDF=pd.DataFrame(sizeSummary)
sizeSummaryDF
```
